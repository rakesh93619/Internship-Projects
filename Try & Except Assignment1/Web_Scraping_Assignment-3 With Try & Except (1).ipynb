### `Question No. 4`:-  Write a python program to search for a smartphone(e.g.: Oneplus Nord, pixel 4A, etc.) on www.flipkart.com and scrape following details for all the search results displayed on 1st page. Details to be  scraped: “Brand Name”, “Smartphone name”, “Colour”, “RAM”, “Storage(ROM)”, “Primary Camera”,  “Secondary Camera”, “Display Size”, “Battery Capacity”, “Price”, “Product URL”. Incase if any of the details is missing then replace it by “- “. Save your results in a dataframe and CSV.

```python
# importing librabries 
import pandas as pd
import numpy as np
import selenium 
from bs4 import BeautifulSoup


from selenium import webdriver

from selenium.common.exceptions import NoSuchElementException, StaleElementReferenceException

import requests
import re
import time

import warnings
warnings.filterwarnings('ignore')
```

```python
driver=webdriver.Chrome('chromedriver.exe')
```

```python
product_url=[]
def smart_phone_name(name):
    driver.get('https://www.flipkart.com/')
    time.sleep(3)
    driver.find_element_by_class_name('_3704LK').send_keys(name)
    driver.find_element_by_xpath('//button[@class="L0Z3Pu"]').click()
    time.sleep(4)
    try:
        tags=driver.find_elements_by_xpath('//a[@class="_1fQZEK"]')
        for i in tags:
            product_url.append(i.get_attribute('href'))
    except NoSuchElementException or StaleElementReferenceException:
        product_url.append('-')
    
```

```python
smart_phone_name('Smartphone')
```

```python
len(product_url)
```

```python
smartphone_name=[]
ram=[]
for i in product_url:
    driver.get(i)
    time.sleep(4)
    try:
        tag=driver.find_element_by_xpath('//span[@class="B_NuCI"]')
        smartphone_name.append(tag.text)
    except NoSuchElementException:
        smartphone_name.append('-') 
    try:
        tag2=driver.find_element_by_xpath('//div[@class="_2418kt"]//li[1]')
        ram.append(tag2.text.split('|')[0])
    except NoSuchElementException:
        ram.append('-') 
```

```python
smartphone_name
```

```python
# As above we can see that the first word of smartphone name is its own brand name so we can scrap the brand name from here
brand_name=[]
for i in smartphone_name:
    brand_name.append(i.split(' ')[0])
brand_name
```

```python
storage=[]
display=[]
front_camera=[]
battery=[]
back_camera=[]
price=[]
for i in product_url:
    driver.get(i)
    time.sleep(4) 
    try:
        tag=driver.find_element_by_xpath('//div[@class="_2418kt"]//li[1]')
        storage.append(tag.text.split('|')[1:])
    except NoSuchElementException:
        storage.append('-')
    try:
        tag2=driver.find_element_by_xpath('//div[@class="_2418kt"]//li[2]')
        display.append(tag2.text)
    except NoSuchElementException:
        display.append('-')
    try:
        tag3=driver.find_element_by_xpath('//div[@class="_2418kt"]//li[3]')
        front_camera.append(tag3.text.split('|')[1])
    except NoSuchElementException:
        front_camera.append('-')
    try:
        tag4=driver.find_element_by_xpath('//div[@class="_2418kt"]//li[4]')
        battery.append(tag4.text)
    except NoSuchElementException:
        battery.append('-')
    try:
        tag5=driver.find_element_by_xpath('//div[@class="_2418kt"]//li[3]')
        back_camera.append(tag5.text.split('|')[0])
    except NoSuchElementException:
        back_camera.append('-')
    try:
        tag6=driver.find_element_by_xpath('//div[@class="_30jeq3 _16Jk6d"]')
        price.append(tag6.text)
    except NoSuchElementException:
        price.append('-')
```

```python
print('length of display : ', len(display))
print('length of battery : ', len(battery))
print('length of front_camera : ', len(front_camera))
print('length of back_camera : ', len(back_camera))
print('length of price : ', len(price))
print('length of storage : ', len(storage))


```

```python
#creating a dataframe
df=pd.DataFrame()
df['Brand_Name']=brand_name
df['SmartPhone_Name']=smartphone_name
df['Product_URL']=product_url
df['Display']=display
df['Ram']=ram
df['Battery_life']=battery
df['Front_Camera']=front_camera
df['Rare_Camera']=back_camera
df['Price']=price
df['Storage']=storage
df
```
