```python
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
```

```python
driver=webdriver.Chrome('chromedriver.exe')
driver.get('https://github.com/')
time.sleep(5)
### assigning the  driver:
act=ActionChains(driver)
### clicking on explore menu
act.click(driver.find_element_by_xpath('/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/summary')).perform()
### clicking on trending button in explore menu
act.click(driver.find_element_by_xpath('/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/div/ul/li[5]/a')).perform()
```

```python
#### finding all the url for every trending repository
url=[]
url_tags=driver.find_elements_by_xpath('//h1[@class="h3 lh-condensed"]//a')
print('The length of url_tags is : ',len(url_tags))
for  i in url_tags:
    url.append(i.get_attribute('href'))
```

```python
url
```

```python
## extracting repository title from all links
title=[]
for i in url:
    driver.get(i)
    time.sleep(3)
    try:
        tag=driver.find_element_by_xpath('//strong[@class="mr-2 flex-self-stretch"]')
        title.append(tag.text)
    except NoSuchElementException:
        title.append("-")
```

```python
title
```

```python
## extracting from all links
description=[]
contributer=[]
language=[]
for i in url:
    driver.get(i)
    time.sleep(3)
    try:
        tag=driver.find_element_by_xpath('//article[@itemprop="text"]')
        description.append(tag.text)
    except NoSuchElementException:
        description.append("-")
    try:
        tag1=driver.find_element_by_xpath('/html/body/div[4]/div/main/div[2]/div/div/div[3]/div[2]/div/div[4]/div/h2/a')
        contributer.append(tag1.text)
    except NoSuchElementException:
        contributer.append("-")
    try:
        tag2=driver.find_element_by_xpath('//ul[@class="list-style-none"]//li[@class="d-inline"][1]')
        language.append(tag2.text)
    except NoSuchElementException:
        language.append("-")    
```

### Creating a dataframe

```python
github_df=pd.DataFrame({'Repository_title':title,
                       'Description':description,
                       'Contributers':contributer,
                       'Language_used':language})
github_df
```

```python
## Extracting the HR Name
hr_name=[]
role=[]
company_name=[]

skills=[]
try:
    name_tags=driver.find_elements_by_xpath('//span[@class="fl ellipsis"]')
    for i in name_tags:
        hr_name.append(i.text)
except NoSuchElementException:
    hr_name.append('-')
try:
    role_tags=driver.find_elements_by_xpath('//span[@class="ellipsis clr"]')
    for i in role_tags:
        role.append(i.text)
except NoSuchElementException:
    role.append('-')
    
try:
    company_tags=driver.find_elements_by_xpath('//a[@class="ellipsis"]')
    count=1
    for i in company_tags:
        if count%2==0:
            company_name.append(i.text)
        count+=1
except NoSuchElementException:
    role.append('-')

try:
    skill_tags=driver.find_elements_by_xpath('//div[@class="hireSec highlightable"]')
    for i in skill_tags:
        skills.append(i.text)
except NoSuchElementException:
    skills.append('-')
```

```python
### extracting the urls 
hr_url=[]
all_links=driver.find_elements_by_xpath('//p[@class="highlightable"]//a[1]')
for i in all_links:
    hr_url.append(i.get_attribute('href'))
```

```python
hr_url
```

```python
## extracting the location
loc=[]
for i in hr_url:
    driver.get(i)
    time.sleep(4)
    try:
        loc_tags=driver.find_elements_by_xpath('//a[@class="fl ellipsis widLrg"]')
        count=1
        for i in loc_tags:
            if count%2==0:
                loc.append(i.text)
            count+=1
    except NoSuchElementException:
        loc.append('-')
```

```python
len(loc)
```

```python
### creating a dataframe
hr_df=pd.DataFrame({'Recruiter_name':hr_name,
                   'Designation':role,
                   'Company_name':company_name,
                   'Skills':skills,
                   'Location':loc})
```

```python
hr_df
```

```python
driver.close()
```

## `Question 10`:-  Details of Datasets from UCI machine learning repositories.
- Url = https://archive.ics.uci.edu/

You have to find the following details:
- A) Dataset name
- B) Data type
- C) Task
- D) Attribute type
- E) No of instances
- F) No of attribute
- G) Year
###### Note: - from the home page you have to go to the ShowAllDataset page through code.

```python
### assign the chrome driver
driver=webdriver.Chrome('chromedriver.exe')

### fetching a web page
driver.get('https://archive.ics.uci.edu/')
```

```python
element_url=driver.find_elements_by_xpath('//span[@class="whitetext"]/a')
urls=[]
for i in element_url:
    urls.append(i.get_attribute('href'))
    
```

```python
urls
```

```python
## our url is last one
## let's get that url
for i in urls:
    driver.get(i)
```

```python
## Extracting data
## some data is stored in another class. We will extract it later
dataset_name=[]
try:
    all_data=driver.find_elements_by_xpath('//tr[@bgcolor="DDEEFF"]')
    for i in all_data:
        dataset_name.append(i.text.split('\n')[0].replace(' ',''))
except NoSuchElementException:
    dataset_name.append('-')


```

```python
len(dataset_name)
```

```python
## extracting the data type
data_type=[]
try:
    all_data=driver.find_elements_by_xpath('//tr[@bgcolor="DDEEFF"]')
    for i in all_data:
        data_type.append(i.text.split('\n')[1])
except NoSuchElementException:
    data_type.append('-')
```

```python
len(data_type)
```

```python
## extracting the year
year=[]
try:
    all_year=driver.find_elements_by_xpath('//tr[@bgcolor="DDEEFF"]')
    for i in all_year:
        year.append(i.text.split('\n')[-1].replace(' ',''))
except NoSuchElementException:
    year.append('-')
len(year)
```

```python
## extracting the number of attributes
attributes=[]
try:
    all_attributes=driver.find_elements_by_xpath('//tr[@bgcolor="DDEEFF"]')
    for i in all_attributes:
        attributes.append(i.text.split('\n')[-2])
except NoSuchElementException:
    attributes.append('-')
len(attributes)
```

```python
## extracting the attributes_type
attribute_type=[]
try:
    all_attribute=driver.find_elements_by_xpath('//tr[@bgcolor="DDEEFF"]//td[4]')
    for i in all_attribute:
        attribute_type.append(i.text)
except NoSuchElementException:
    attribute_type.append('-')
len(attribute_type)
```

```python
## extracting the task type
task=[]
try:
    all_task=driver.find_elements_by_xpath('//tr[@bgcolor="DDEEFF"]//td[3]')
    for i in all_task:
        task.append(i.text)
except NoSuchElementException:
    task.append('-')
len(task)
```

```python
## extracting the No of intances
intances=[]
try:
    all_attribute=driver.find_elements_by_xpath('//tr[@bgcolor="DDEEFF"]//td[5]')
    for i in all_attribute:
        intances.append(i.text)
except NoSuchElementException:
    intances.append('-')
len(intances)
```

```python
### creating a dataframe
dataset_df=pd.DataFrame({'Dataset_name':dataset_name,
                        'Data_type':data_type,
                        'Dafault_task':task,
                        'Attribute_type':attribute_type,
                        'No_of_intances':intances,
                        'No_of_attributes':attributes,
                        'Year':year})
```
